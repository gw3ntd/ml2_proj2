```python
import tensorflow as tf
import keras
import numpy as np
import pandas as pd
from customlayer import MyLayerNorm
from functions import get_keras_vals, get_custom_vals, comp_layer
import matplotlib.pyplot as plt
```

```python
(train_images, train_labels), (test_images, test_labels) = tf.keras.datasets.fashion_mnist.load_data()
```

```python
train_images = train_images / 255.0
test_images = test_images / 255.0
```

```python
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
```

```python
train_images.shape
```

```python
plt.figure()
plt.imshow(train_images[1])
plt.colorbar()
plt.grid(False)
plt.show()
```

```python
plt.figure(figsize=(10,10))
for i in range(25):
    plt.subplot(5,5,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(train_images[i], cmap=plt.cm.binary)
    plt.xlabel(class_names[train_labels[i]])
plt.show()
```

```python
mnist_df = comp_layer(tensor=train_images, batch_size=100)
```

```python
plt.plot(mnist_df[0], label="L2 Norm", color="orange")
plt.plot(mnist_df[1], label="L INF Norm", color="green")
plt.title("Comparison Between Keras LayerNormalization and \nCustom LayerNormalization - Fashion MNIST")
plt.xlabel("Batch Index")
plt.ylabel("Value")
plt.legend()
plt.show()
```
